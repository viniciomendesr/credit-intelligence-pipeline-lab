```python
df = pd.read_csv(data / "cs-training.csv", index_col=0)
```

```python
print("\nshape:", df.shape)
print("\nTipos:\n", df.dtypes)

# print("\nTestando df nulos\n", df.isnull())
# print("\nTestando somatória de df nulos\n", df.isnull().sum())

# Nulos por coluna
null_pcoluna = (df.isnull().sum() / len(df) * 100).round(2)
print("\nNulos (%):\n", null_pcoluna[null_pcoluna > 0])

# Valores impossíveis

idades_invalidas = (df["age"] <= 0).sum()
salarios_invalidos = (df["MonthlyIncome"] <= 0).sum()

print("\n Idades Invalidas: ", idades_invalidas)
print("Salarios Invalidas: ", salarios_invalidos)


# Taxa de inadiplência
defaut_rate = df["SeriousDlqin2yrs"].mean()
print("\n Taxa de Default: ", defaut_rate)

# Distruibuição de renda

dist_renda = df["MonthlyIncome"].quantile([0.25, 0.5, 0.75, 0.99]);
print("\n Distribuicao de Renda: ", dist_renda)
```
